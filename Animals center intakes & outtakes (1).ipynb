```python
# Task 01: Print the shape of the intake dataset
intake_dataset = pd.read_csv('DataSet_Animal_Center_Intakes.csv')
print("Intake Dataset Shape:", intake_dataset.shape)
```

```python
# Task 02: Identify columns with null values in the intake dataset
columns_with_null_values = intake_dataset.columns[intake_dataset.isnull().any()].tolist()
print("Columns with null values in the intake dataset:", columns_with_null_values)
```

```python

# Task 03: Remove duplicate rows from the intake dataset
intake_dataset.drop_duplicates(inplace=True)
```

```python
# Task 04: Rename column names in the intake dataset
intake_dataset.rename(columns={
    'Animal ID': 'id',
    'DateTime': 'date_time',
    'MonthYear': 'month_year',
    'Found Location': 'found_location',
    'Intake Type': 'intake_type',
    'Intake Condition': 'intake_condition',
    'Animal Type': 'animal_type',
    'Sex upon Intake': 'sex_upon_intake',
    'Age upon Intake': 'age_upon_intake',
    'Breed': 'breed'
}, inplace=True)

```

```python
# Task 05: Remove the 'MonthYear' column from the intake dataset
intake_dataset.drop('month_year', axis=1, inplace=True)

```

```python
# Task 06: Extract month and year from the 'DateTime' column in the intake dataset
intake_dataset['month'] = pd.to_datetime(intake_dataset['date_time']).dt.month
intake_dataset['year'] = pd.to_datetime(intake_dataset['date_time']).dt.year

```

```python
# Task 07: Generate a pie chart representing year-wise animal intake from 2014 to 2021
yearly_intake_counts = intake_dataset[intake_dataset['year'].between(2014, 2021)]['year'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(yearly_intake_counts, labels=yearly_intake_counts.index, autopct='%1.1f%%')
plt.title('Yearly Animal Intake Distribution (2014-2021)')
plt.show()
```

```python
# Task 08: Generate a bar chart representing month-wise animal intake
monthly_intake_counts = intake_dataset['month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(monthly_intake_counts.index, monthly_intake_counts.values)
plt.xlabel('Month')
plt.ylabel('Animal Intake Count')
plt.title('Monthly Animal Intake')
plt.xticks(range(1, 13))
plt.show()
```

```python
# Task 09: Generate a bar chart representing animal intake method-wise counting
intake_method_counts = intake_dataset['intake_type'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(intake_method_counts.index, intake_method_counts.values)
plt.xlabel('Intake Method')
plt.ylabel('Count')
plt.title('Animal Intake Methods')
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 10: Print the unique animal intake conditions
animal_conditions = intake_dataset['intake_condition'].unique()
print("Animal Intake Conditions:")
for condition in animal_conditions:
    print(condition)
```

```python
# Task 11: Print the types of animals and their counts in the intake dataset
animal_types_count = intake_dataset['animal_type'].value_counts()
print("Animal Types and Counts:")
print(animal_types_count)
```

```python
# Task 12: Generate a bar plot representing the top 10 breeds in the intake dataset
top_10_breeds = intake_dataset['breed'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.bar(top_10_breeds.index, top_10_breeds.values)
plt.xlabel('Breed')
plt.ylabel('Count')
plt.title('Top 10 Breeds in Animal Shelter')
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 13: Generate a bar plot representing the top 20 common names of animals in the intake dataset
top_20_names = intake_dataset['Name'].value_counts().head(20)
plt.figure(figsize=(12, 6))
plt.bar(top_20_names.index, top_20_names.values)
plt.xlabel('Name')
plt.ylabel('Count')
plt.title('Top 20 Common Names of Animals in Animal Shelter')
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 14: Generate a bar chart representing the "Outcome Type" wise count of animals using the outcome dataset
outcome_dataset = pd.read_csv('DataSet_Animal_Center_Outcomes.csv')
outcome_type_counts = outcome_dataset['Outcome Type'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(outcome_type_counts.index, outcome_type_counts.values)
plt.xlabel('Outcome Type')
plt.ylabel('Count')
plt.title('Outcome Type-wise Count of Animals')
plt.xticks(rotation=45)
plt.show()
```
